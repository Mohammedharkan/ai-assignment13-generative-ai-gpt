## GPT Architecture

Generative Pre-trained Transformers (GPTs) are based on the Transformer architecture, which uses a self-attention mechanism to understand relationships between words in a sentence. Unlike older sequential models, transformers process all input tokens simultaneously, making training faster and more efficient.

The text generation process starts with tokenization, where the input text is divided into tokens and converted into numerical IDs. These tokens are passed through the GPT model, which predicts the probability of the next token based on the previous context. This process is repeated one token at a time until the desired output length is reached.

GPT models are pre-trained on very large text datasets, allowing them to learn grammar, language patterns, and contextual relationships. During text generation, parameters such as temperature, Top-k, and Top-p influence how the next token is selected, producing outputs that range from predictable to creative.

```python
# Import required libraries

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

## 2. Model and Tokenizer Loading

In this section, I load a pre-trained GPT-2 model and its tokenizer from the Hugging Face Transformers library. The tokenizer converts text into numerical tokens, while the model uses these tokens to generate meaningful text.

```python
# Load GPT-2 tokenizer and model

model_name = "distilgpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)

model = GPT2LMHeadModel.from_pretrained(model_name)

print("Model loaded successfully!")
print("Tokenizer loaded successfully!")
```

## 3. Text Generation Function

In this section, I create a reusable function that generates text using the pre-trained GPT model. The function takes a text prompt as input and returns generated text based on different generation parameters.

```python
# Text Generation Function

def generate_text(prompt,
                  max_length=50,
                  temperature=1.0,
                  top_k=50,
                  top_p=0.95):

    # Convert text into tokens
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate text
    outputs = model.generate(
        **inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Convert tokens back to text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text
```

## 4. Exercise 1: Basic Text Generation

In this exercise, I use the pre-trained GPT model to generate text from a simple prompt. This demonstrates the model's ability to continue a sentence based on the context provided.

```python
# Exercise 1: Basic Text Generation

prompt = "Artificial Intelligence is"

generated_text = generate_text(prompt)

print("Prompt:")
print(prompt)

print("\nGenerated Text:")
print(generated_text)
```

### Observation

The model successfully generated text based on the prompt. The output is related to Artificial Intelligence, but it stopped before completing the final sentence because the default maximum length was limited. This shows how output length affects text generation.

## 5. Exercise 2: Controlling Output Length

In this exercise, I compare text generation using two different output lengths. This demonstrates how the maximum length parameter affects the amount of generated text.

```python
# Exercise 2: Controlling Output Length

prompt = "Artificial Intelligence is"

print("========== Short Output ==========\n")
print(generate_text(prompt, max_length=30))

print("\n\n========== Long Output ==========\n")
print(generate_text(prompt, max_length=100))
```

### Observation

Increasing the value of `max_length` allows the model to generate longer and more detailed text. A smaller value produces shorter responses, while a larger value gives the model more space to continue the generated content.

## 6. Exercise 3: Adjusting Temperature

In this exercise, I compare text generation using different temperature values. Lower temperatures produce more predictable text, while higher temperatures create more diverse and creative outputs.

```python
# Exercise 3: Adjusting Temperature

prompt = "Artificial Intelligence is"

print("========== Temperature = 0.3 ==========\n")
print(generate_text(prompt, temperature=0.3))

print("\n\n========== Temperature = 1.0 ==========\n")
print(generate_text(prompt, temperature=1.0))

print("\n\n========== Temperature = 1.5 ==========\n")
print(generate_text(prompt, temperature=1.5))
```

### Observation

Lower temperature values produce more predictable and focused text. Higher temperature values generate more creative and diverse responses, but the text may become less consistent.

## 7. Exercise 4: Using Top-k and Top-p

In this exercise, I compare text generation using different values of Top-k and Top-p. These parameters control how the model selects the next word by limiting the candidate words based on either the number of choices or the cumulative probability.

```python
# Exercise 4: Using Top-k and Top-p

prompt = "Artificial Intelligence is"

print("========== Top-k = 20 ==========\n")
print(generate_text(prompt, top_k=20))

print("\n\n========== Top-k = 80 ==========\n")
print(generate_text(prompt, top_k=80))

print("\n\n========== Top-p = 0.70 ==========\n")
print(generate_text(prompt, top_p=0.70))

print("\n\n========== Top-p = 0.95 ==========\n")
print(generate_text(prompt, top_p=0.95))
```

### Observation

Lower Top-k and Top-p values produce more focused and predictable text because the model selects from fewer candidate words. Higher values increase diversity by allowing the model to consider a wider range of possible words during text generation.

## 8. Exercise 5: Prompt Engineering

In this exercise, I use different prompts to demonstrate how the quality and style of the generated text depend on the prompt provided to the model.

```python
# Exercise 5: Prompt Engineering

prompts = [
    "Write a short story about a robot.",
    "Explain machine learning in simple words.",
    "Write a professional email to a customer about a delayed order."
]

for i, prompt in enumerate(prompts, 1):
    print(f"\n========== Prompt {i} ==========")
    print("Prompt:")
    print(prompt)

    print("\nGenerated Text:")
    print(generate_text(prompt))

    print("\n" + "="*60)
```

### Observation

Different prompts produce different types of responses. Clear and specific prompts help the model generate more relevant and useful text, demonstrating the importance of prompt engineering in Generative AI applications.

Generative AI has many benefits, but it also raises ethical concerns. These models may generate incorrect information, biased content, or text that sounds realistic but is not true. They may also be misused to create spam, fake news, or misleading content.
